```python
sns.pairplot(parkinsons_data, vars=['MDVP:Fo(Hz)', 'MDVP:Fhi(Hz)', 'MDVP:Flo(Hz)', 'MDVP:Jitter(%)'], hue='status')
plt.title('Pairplot of Selected Features')
plt.show()
```

```python
sns.boxplot(x='status', y='MDVP:Shimmer', data=parkinsons_data)
plt.title('Boxplot of MDVP:Shimmer by Status')
plt.show()
```

```python
# Ensure only numeric columns are selected for the mean calculation
numeric_columns = parkinsons_data.select_dtypes(include=[float, int])

# Now perform the groupby operation based on the 'status' column
grouped_data = numeric_columns.groupby(parkinsons_data['status']).mean()

# Display the result
print(grouped_data)
```

```python
plt.figure(figsize=(12, 10))
correlation_matrix = grouped_data.corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```
